```python
import pandas as pd
import numpy as np
```

```python
def read_cancer_data(file = 'main'):
    
    
    main_link = ('data/'
                 +'https://www.england.nhs.uk/'
                 +'statistics/wp-content/uploads/'
                 + 'sites/2/2023/12/'+
                 'CWT-CRS-2022-23-Data-Extract-Provider-Final.xlsx')
    
    if file == 'main':
        data = pd.read_excel(main_link)
        
    return data
```
